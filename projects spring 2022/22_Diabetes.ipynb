## Relationship Between Education vs Diabetes

```python
Education_vs_Diabetes= diabetes_data.pivot_table('Diabetes_012',index='Diabetes_type',columns='Education',aggfunc='count')
Education_vs_Diabetes
```

```python
fig, ax = plt.subplots(figsize=(15,5)) 
BMI_vs_Diabetes1= sns.histplot(data=diabetes_data,x='Diabetes_type',hue='Education',multiple='dodge',stat='count',bins=20,palette= 'Blues',shrink=0.5)
BMI_vs_Diabetes1.set_yscale('log')
BMI_vs_Diabetes1.axhline(10**4,color= "red", linestyle = '--')
plt.title('Education Vs Diabetes')
```

#### Education Dictionary:

##### levels of Education: 

###### 1 = Never attended school or only kindergarten
###### 2 = Grades 1 through 8 (Elementary) 
###### 3 = Grades 9 through 11 (middle school) 4 
###### 4 = Grade 12 or GED (High school graduate)
###### 5 = College 1 year to 3 years 6
###### 6 = College 4 years or more (College graduate)

### From the above we can see that on the one hand, the number of non-diabetic people increases in line with the level of education along the way, and on the other hand we have the diabitic and pre-diabitic groups whose level of education raising until level 4 and after it is on a downward trend.

### As a result, we can conclude that diabetic and pre-diabetic people have difficulty in acquiring higher education than healthy people. 

### As a recommendation we offer to take into account as a criterion not only the socio-economic status of the student for receiving scholarships in higher education, But also his state of health.

## Relationship Between BMI and Diabetes

```python
Diabetes_vs_BMI= diabetes_data.pivot_table('Diabetes_012',index='BMI',columns='Diabetes_type',aggfunc='count')
Diabetes_vs_BMI.fillna(0)
```

```python
fig, ax = plt.subplots(figsize=(20,10)) 
BMI_vs_Diabetes= sns.histplot(data=diabetes_data,x='BMI',hue='Diabetes_type',multiple = 'stack',stat='count')
BMI_vs_Diabetes.set_xlim(0,70)
plt.title('BMI vs Diabetes')
```

```python
Diabetes_vs_BMI_sliced = Diabetes_vs_BMI.iloc[:, 0:1]
Diabetes_vs_BMI_sliced

Diabetes_vs_BMI_sliced.plot(kind = "line",color = "Blue")
plt.ylim(0,5000)
plt.xlim(15,60)
plt.axvline(x=25,color= "Red",linestyle='--')
plt.axvline(x=34,color= "Red",linestyle='--')
plt.ylabel('Number of Diabetic people')
plt.title("BMI vs Only Diabetic people")
```

### We can infer from the fig above that Diabetic people are more likely to have BMI range between 25 to 34 and suffer from Overweight and obesity. As a result they must be under close supervision and perform periodic inspections on their weight. In addition, the World Diabetes Foundation must take action by producing training and nutrition programs.
