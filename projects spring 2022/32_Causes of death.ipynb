 <div dir='rtl'>         2019             

```python
year2019= death_df.loc[(death_df.Year == 2019)].copy()
```

```python
df_new =year2019.sum(axis=0).reset_index()
df_new=df_new.rename(columns={'index': 'cause', 0:'total deaths in dozensd of millions'})
df_new.drop(df_new.index[[0,1 , -1]],inplace=True)
df_new.convert_dtypes()
df_new = df_new.sort_values(by=["total deaths in dozensd of millions"],ascending=False)
```

```python
fig4, ax = plt.subplots(figsize =(15,3))
sns.barplot(x='cause', y='total deaths in dozensd of millions', data=df_new)
sns.set_theme(style="darkgrid")
plt.xticks(rotation=90)
plt.ylabel('total deaths in dozensd of millions',size=20)
plt.xlabel('cause',size=20)
plt.title('Top Death Causes In The World- 2019',size=30, fontweight = 'bold')
plt.show()
```

```python
perbyyear.drop(per_pop.loc[perbyyear.populaion == 0].index, inplace = True)
```
