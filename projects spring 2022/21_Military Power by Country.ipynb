```python
war= army[(army['country']== 'Russia') | (army['country']== 'Ukraine')|(army['country']== 'Israel')].copy() 
```

### <div dir="rtl">רצינו לחקור על המלחמה בין רוסיה לאוקראינה. בחרנו להציג פרמטרים כמותיים שיציגו את יחסי הכוחות בין המדינות בנוסף לצורך "קנה מידה" הוספנו את מדינת ישראל. 


```python
fig,axes= plt.subplots(figsize= (20,10),nrows=2 , ncols= 3 )
sns.barplot(ax= axes[0,0], x= 'country' ,y='Tanks', data= war,palette= 'summer'); 
sns.barplot(ax= axes[0,1], x= 'country' ,y='Total_Aircraft_Strength', data= war,palette= 'cool_r');
sns.barplot(ax= axes[0,2], x= 'country' ,y= 'Navy_Ships', data= war,palette= 'hot_r');
sns.barplot(ax= axes[1,0], x= 'country' ,y='Country_Budget', data= war,palette= 'summer');
sns.barplot(ax= axes[1,1], x= 'country' ,y= 'Active_Personnel', data= war,palette= 'cool_r');
sns.barplot(ax= axes[1,2], x= 'country' ,y= 'Attack_Helicopters', data= war,palette= 'hot_r');
```

### <div dir="rtl">מהגרפים לעיל, ניתן לראות כי בכלל הפרמטרים שנבדקו, רוסיה ביתרון חד משעמי לעומת ישראל ואוקראינה 
[מצב מלחמה עדכני רוסיה-אוקראינה](https://makospecial.co.il/ukrainewar)

```python
war['perc_active_available']= army['Active_Personnel']/army['Available_Manpower']*100
war['perc_aircraft_area']= army['Total_Aircraft_Strength']/army['Square_Land_Area']*100
```

יצרנו עמודות של יחס בין כמות חיילים פעילים לבין כמות אנשים שכשירים לשרת בצבא
בנוסף עמודה של יחס בין כח אווירי ביחס לגודל המדינה

### <div dir="rtl"><span style='background :yellow' > :לאחר שהצגנו את הנתונים הכמותיים, כעת נציג נתונים יחסיים באחוזים </span>
<div dir="rtl">1. יחס בין מספר החיילים לבין מספר אנשים כשירים לשירות
<div dir="rtl">2. יחס בין עוצמת הכח האווירי לבין שטח המדינה
<div dir="rtl">3. יחס בין תקציב המדינה לבין תקציב הביטחון



```python
fig,axes= plt.subplots(figsize= (15,5) , ncols= 3 )
sns.barplot(ax= axes[0], x= 'country' ,y='perc_active_available', data= war).set(title= "תורישל םירישכ םילייח רפסמ");
sns.barplot(ax= axes[1], x= 'country' ,y='perc_def_budget', data= war, ).set(title= "ןוחטיבה ביצקתל הנידמה ביצקת");
sns.barplot(ax= axes[2], x= 'country' ,y= 'perc_aircraft_area', data= war).set(title= "הנידמה חטשל יריווא חוכ");
```

#### <div dir="rtl">מהניתוח שערכנו ניתן לראות כי רוסיה ביחס לישראל ואוקראינה אכן בעלת הכח הצבאי ההרב ביותר מבחינה כמותית, מה שמאשש את יתרונה במלחמה.  עם זאת ניתן להבחין כי ישראל משקיעה את האחוז הגבוה ביותר בהשקעה כספית, הקצאת כח אדם ומשאבים צבאיים.

###  <div dir="rtl">  במידה ואוקראינה הייתה מתנהלת כמו ישראל ומגדילה את משאביה ביטחוניים, כמו הגדלת צי המטוסים ויותר משרתים פעילים, אולי מצבה היה אחר מהקיים בימנו אנו.



## <div dir="rtl"> כעת נציג מפת חום, ביבשת אירופה בלבד, על עמודות נבחרות מהמידע:

```python
features= ['External_Debt','Country_Budget','Defense_Budget','Active_Personnel','Attack_Helicopters','Square_Land_Area','Tanks','Total_Aircraft_Strength' ,'Total_Population','Navy_Ships','Shared_Borders']
corrs = army[army.Mainland == 'Europe'][features].corr(method = 'spearman')

plt.figure(figsize=(25,20))  
sns.heatmap(corrs, cmap='coolwarm', center=0, annot = True);
```

##### <div dir="rtl">השתמשתנו בקורלציית פירסון כי ראינו שהדאטה לא מתנהגת נורמלית ברובה לכן השתמשנו בקורלציה זו

##### <div dir="rtl"><span style='background : yellow' > ממפת החום ניתן לראות קורלציה של 0.89 בין כח אווירי לבין תקציב הבטחון

```python
sns.regplot(data= army[army.Mainland == 'Europe'] , y= 'Defense_Budget', x= 'Total_Aircraft_Strength')
```

#### <div dir="rtl"> לאחר שמצאנו התאמה בין השקעה בתקציב הביטחון לבין השקעה בכח אווירי, נבחן את זה כנגד טנקים וספינות חייל הים במדינות אירופה שאנו מעריכים כמדינות מפותחות 

```python
features= ['Defense_Budget','Tanks','Total_Aircraft_Strength' ,'Navy_Ships']
corrs = army[army.Mainland == 'Europe'][features].corr(method = 'spearman')

plt.figure(figsize=(5,5))  
sns.heatmap(corrs, cmap='coolwarm', center=0, annot = True);
```

```python
fig, axes = plt.subplots(figsize=(10, 5), ncols=3)
Europe= army[army.Mainland == 'Europe']
sns.regplot(data= Europe[Europe.Total_Aircraft_Strength<4000] , y= 'Defense_Budget', x= 'Total_Aircraft_Strength',ax= axes[0])
sns.regplot(data= Europe[Europe.Navy_Ships <600] , y= 'Defense_Budget', x= 'Navy_Ships',ax= axes[1], color= 'r')
sns.regplot(data= Europe[Europe.Tanks <12000], y= 'Defense_Budget', x= 'Tanks', ax= axes[2],color='g')
plt.tight_layout()
```
